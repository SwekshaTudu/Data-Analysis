```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv('C:\\Users\\sweks\\Downloads\\householdtask3.csv')
```

```python
display(data.head(10))
```

```python
#Scatter plot with your against own
plt.scatter(data['year'], data['own'])

#additle title to the plot
plt.title('Scatter Plot')

#setting the x and y label
plt.xlabel('Year')
plt.ylabel('Own')
plt.show()

```

```python
#line chart with year againstm own
plt.plot(data['year'])
plt.plot(data['own'])
plt.title('Line Chart')
plt.xlabel('year')
plt.ylabel('own')
plt.show()

         
```

```python
#bar chart or bar plot
plt.bar(data['year'], data['own'])
plt.title("Bar plot")
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
#histogram
plt.hist(data['income'])
plt.title("Histogram")
plt.show()
```
