```python
questions = ["1. How many people have heart disease and how many people doesn't havr heart disease?"
             "2. People of which sex has most heart disease?"
              "3. People of which sex has whih tyoe of chest pain most?",
             "4. People with which chest pain are most pron to have heart disease?"
             "5. What is the relationship between maximum heart rate achieved (thalach) and heart disease",
             "6. Does exercise-induced angina (exang) correlate with the likelihood of heart disease?",
             "7. How does ST depression induced by exercise (oldpeak) relate to heart disease risk?",
             "8. What is the significance of the slope of the peak exercise ST segment (slope) in predicting heart disease?",]
questions
            
```

```python
#question 1
df.target.value_counts()
```

```python
#question 2
df.target.value_counts().plot(kind = "bar", color = ["orchid", "salmon"])
plt.title("Heart disease value")
plt.xticks(rotation = 0)
plt.xlabel("1= Heart Disease, 0 = No heart disease")
plt.ylabel("Amount");
```

```python
#plotting a pie chart
df.target.value_counts().plot( kind= 'pie', figsize = (8,6))
plt.legend(["Disease", " No disease"]);
```

```python
df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind = 'pie', figsize= (8,6))
plt.title('Male Female Ratio')
plt.legend(['Male', 'Female']);
```

```python
#2nd question 
pd.crosstab(df.target , df.sex)
```

```python
sns.countplot(x= 'target' , data =df, hue = 'sex')
plt.title("Heart Disease Frequency for sex")
plt.xlabel("0 = No heart Disease, 1 = Heart Disease");
```

```python
#3 people of which sex has which type of chest pain most
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind = 'bar', color = ['salmon', 'lightskyblue', 'springgreen', 'khaki'])
plt.title('Chest pain type vs count');
plt.xticks(rotation=0)
```

```python
pd.crosstab(df.sex, df.cp)
```

```python
pd.crosstab(df.sex, df.cp).plot(kind = 'bar', color= ['coral', 'lightskyblue', 'plum', 'khaki'])
plt.title('Type of chest pain for sex')
plt.xlabel('0 = Female, 1= Male');
```

```python
#people with which chest pain are most pron to heart disease?
pd.crosstab(df.cp, df.target)
```

```python
sns.countplot(x= 'cp', data = df, hue= 'target');
```

```python
#5th question maximum heart rate achieved and heart disease

sns.boxplot(data = df, x='target', y='thalach')
plt.title('Max heart rate vs heart disease')
```

```python
#6th Exercise induced angina and heart disease
sns.countplot(data=df, x='exang',hue ='target')
plt.title('exercise indued angina vs heart disease')
```

```python
# ST depressiion induced by exercise and heart disease
sns.boxplot(data = df, x='target', y='oldpeak')
plt.title(' ST Depression vs Heart Disease')
```

```python
#slope of peak exercise ST Segment and Heart Disease
sns.countplot(data=df,  x= 'slope', hue= 'target')
plt.title('Slope of peak exercise ST segment vs heart disease')
```
